```python
us_names = pd.read_csv("Name origins - US.tsv", sep='\t')
us_names.head()
```

```python
from vega_datasets import data

us_names1 = us_names.iloc[:,:3]
states = alt.topo_feature(data.us_10m.url, feature='states')
pop = data.population_engineers_hurricanes()


outlines = alt.Chart(states, title='Origins of state names').mark_geoshape(
        stroke='black'
    ).encode(
        color=alt.Color('Named after:N', scale=alt.Scale(scheme='category20'))
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(pop, 'id', ['state'])
    ).transform_lookup(
        lookup='state',
        from_=alt.LookupData(us_names1, 'State', ['Named after','Origin'])
    ).project('albersUsa').properties(
        width=800,
        height=500
    )
outlines
```
